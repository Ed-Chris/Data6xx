```python
# Import first dataset 

import pandas as pd
temperaturedata = pd.read_csv("GlobalLandTemperaturesByCountry.csv")
temperaturedata

# Change the dt column to year
temperaturedata = temperaturedata.rename(columns={"dt": "Year"})
temperaturedata

# Edit the Year column and combine 
temperaturedata['Year'] = temperaturedata['Year'].str.split('-').str[0]

# Convert the "Year" column to integers
temperaturedata['Year'] = temperaturedata['Year'].astype(int)

# Find any missing values
missing_values = temperaturedata.isna()
missing_values 

# Drop missing values
temperaturedata = temperaturedata.dropna()
temperaturedata 

# Filter only the years from 2000 to 2013 
filtered_temperaturedata = temperaturedata[(temperaturedata['Year'] >= 2000) & (temperaturedata['Year'] <= 2013)]
filtered_temperaturedata.head(60)

# Search specific country
specific_countries = filtered_temperaturedata[filtered_temperaturedata['Country'] == 'Afghanistan']
specific_countries                        
```

```python
# Import second dataset

import pandas as pd
productiondata = pd.read_csv("modern-renewable-prod.csv")
productiondata

# Rename the Entity column to "Country"
productiondata = productiondata.rename(columns={"Entity": "Country"})

# Drop the Code column 
productiondata.drop('Code', axis=1, inplace=True)
productiondata

# Filter only the years from 2000 to 2013
filtered_productiondata = productiondata[(productiondata['Year'] >= 2000) & (productiondata['Year'] <= 2013)]
filtered_productiondata.head(30)

# Find any missing values
missing_values = filtered_productiondata.isna()
missing_values

# Drop the missing values 
cleaned_productiondata = filtered_productiondata.dropna()
cleaned_productiondata.head(30)
```

```python
# Import third dataset

import pandas as pd
consumptiondata = pd.read_csv("World Energy Consumption.csv")
consumptiondata

# Unique columns in the dataset
unique_columns_list = list(consumptiondata.columns)
unique_columns_list

# Select the columns of interest
consumptiondata = consumptiondata[['country', 'year', 'fossil_fuel_consumption', 'wind_consumption', 'hydro_consumption', 'solar_consumption', 'biofuel_consumption']]
consumptiondata

# Filter only the years from 2000 to 2013
filtered_consumptiondata = consumptiondata[(consumptiondata['year'] >= 2000) & (consumptiondata['year'] <= 2013)]
filtered_consumptiondata.head(30)

# Find any missing vlaues
missing_values = filtered_consumptiondata.isna()
missing_values

# Drop na columns
cleaned_consumptiondata = filtered_consumptiondata.dropna()
cleaned_consumptiondata

# Rename the columns
cleaned_consumptiondata = cleaned_consumptiondata.rename(columns={"country": "Country", "year": "Year"})
cleaned_consumptiondata.head(30)

```

```python
# Import fourth dataset

import pandas as pd
gasdata = pd.read_csv("co2-ghg.csv")
gasdata

# Unique columns in the dataset
unique_columns_list = list(gasdata.columns)
unique_columns_list

# Select the columns of interest
gasdata = gasdata[['country', 'year', 'co2', 'coal_co2', 'total_ghg', 'methane', 'nitrous_oxide', 'population']]
gasdata

# Filter only the years from 2000 to 2013
filtered_gasdata = gasdata[(gasdata['year'] >= 2000) & (gasdata['year'] <= 2013)]
filtered_gasdata.head(30)

# Find any missing vlaues
missing_values = filtered_gasdata.isna()
missing_values

# Drop na columns
cleaned_gasdata = filtered_gasdata.dropna()

# Rename columns
cleaned_gasdata = cleaned_gasdata.rename(columns={"country": "Country", "year": "Year","population": "Population"})
cleaned_gasdata.head(30)

cleaned_gasdata.tail(30)
```

```python
# Merge the datasets based on "Year" and "Country" columns
merged_data = pd.merge(filtered_temperaturedata, cleaned_productiondata, on=["Year", "Country"], how="inner")
merged_data = pd.merge(merged_data, cleaned_gasdata, on=["Year", "Country"], how="inner")
merged_data = pd.merge(merged_data, cleaned_consumptiondata, on=["Year", "Country"], how="inner")

merged_data

# Convert the merged data to a csv
# merged_data.to_csv("merged_data.csv", index=False)
```
